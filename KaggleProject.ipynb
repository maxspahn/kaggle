Importing the necessary modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

Loading the data storing the criteria

```python
# @todo the date is not yet taken into account
X = np.loadtxt('train.csv',  delimiter=',', usecols=range(2,14),unpack = True, skiprows=1)

#print X[0],X[5]
print len(X[0])
```

```python
# Load the number of bikes rent, 'counts'
# @todo how to deal with the counts as label? Multiple classes?
y = np.loadtxt('train.csv', delimiter=',', usecols=[14], dtype='int',unpack = True, skiprows=1)

# Python 3 alternative:
#y = np.loadtxt('test.csv', delimiter=',',skiprows=1, usecols=[3001], dtype='bytes').astype('str')


# At the moment there is a seperation in 5 classes

y = np.where(y>800, 5 ,y)
y = np.where(y>600, 4, y)
y = np.where(y>400, 3, y)
y = np.where(y>200, 2, y)
y = np.where(y>5, 1, y)

#print y
```

```python
# Load pylab
%pylab inline
```

```python
# Visualize the data across 2 of its axes
idx_1 = 8 # first axis
idx_2 = 10 # second axis
plt.scatter(X[idx_1,y==0], X[idx_2,y==0], # y==0 gives the indices of the columns for which y is equal to 0
            color='yellow', label='1')
plt.scatter(X[idx_1,y==1], X[idx_2,y==1], color='green', label='2')
plt.scatter(X[idx_1,y==2], X[idx_2,y==2], color='blue', label='3')
plt.scatter(X[idx_1,y==3], X[idx_2,y==3], color='orange', label='4')
plt.scatter(X[idx_1,y==4], X[idx_2,y==4], color='red', label='5')
plt.legend(scatterpoints=1)
plt.xlabel('temp %d' % idx_1, fontsize=14)
plt.ylabel('windspeed %d' % idx_2, fontsize=14)
```
